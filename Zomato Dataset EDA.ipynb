```python
final_df = pd.merge(df, df_Country, on='Country Code', how='left') 
# on means on which column we need to merge the two dataframes.
# how means how we need to merge the two dataframes i.e, left, right, inner, outer. So here we are using left join i.e, it will be added to the df (left) as df_Country has only 2 columns.
```

# Finding Out Countries with 0 Rating

```python
# Countries with 0 rating
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index().rename(columns={0:'0 Rating Count'})
```

```python
# .groupby('Country') means we are grouping the data based on the countries.
# .size() means we are counting the number of rows in the dataset.
# .groupby('Country').size() means we are grouping the data based on the countries and counting the number of rows in the dataset.
```

```python
## Other way to do the same thing.
final_df.groupby(['Aggregate rating','Country']).size().reset_index().head(4).rename(columns={'Aggregate rating':'Average Rating',0:'0 Rating Count'})
```

# Observation
The most of 0 rating is from India following Brazil, United States and United Kingdom
